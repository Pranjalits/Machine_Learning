```python
#html text is passed in beautiful soup and soup is created out of it.
#soup is just an object where we have the whole strength and we can perform multiple operations on this.
soup = BeautifulSoup(html,'html.parser')
print(soup)
```

```python
data = []
for quote in soup.find_all('div',class_ = 'quote'):
    q = quote.find('span', class_ = 'text').text[1:-1]
    a = quote.find('small', class_ = 'author').text
    id = quote.find('a').get('href')
    tags = []
    for tag in quote.find_all('a',class_ = 'tag'):
        tags.append(tag.text)
    tags = ','.join(tags)
    data.append([q,a,id,tags])

data[0]
```

```python
from tqdm import tqdm
for page in tqdm(range(2,11)) :
  link = 'https://quotes.toscrape.com/page/' + str(page)
  response = requests.get(link)
  # This line sends an HTTP GET request to the URL stored in link to retrieve the content of the web page.
  soup = BeautifulSoup(response.text, 'html.parser')
  for quote in soup.find_all('div',class_ = 'quote'):
      q = quote.find('span', class_ = 'text').text[1:-1]
      a = quote.find('small', class_ = 'author').text
      id = quote.find('a').get('href')
      tags = []
      for tag in quote.find_all('a',class_ = 'tag'):
          tags.append(tag.text)
      tags = ','.join(tags)
      data.append([q,a,id,tags])
```
